Для начала обсудим условия решения задачи:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def process_work_experience(raw_string: str) -> int:
    """Calculates total work experience in month

    Args:
        raw_string (str): input work experience description

    Returns:
        int: work experience in month
    """
    
    if raw_string is np.nan or raw_string == 'Не указано':
        return np.nan
    
    total_month = 0

    for pattern in (' месяц', ' лет', ' год'):

        if pattern not in raw_string[:30]:
            continue

        string_split = raw_string[:30].split(pattern)[0] 
        
        n_month = int(string_split.split()[-1])
                
        if pattern in (' лет', ' год'):
            n_month *= 12

        total_month += n_month
        
    return total_month
    

proc_df['Опыт работы (месяц)'] = proc_df['Опыт работы'].apply(process_work_experience)

proc_df = proc_df.drop(['Опыт работы'], axis=1)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь

def get_city(raw_string: str) -> str:
    """Get city name from description

    Args:
        raw_string (str): input string

    Returns:
        str: city name
    """
    
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 
                  'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск',
                  'Пермь', 'Воронеж','Волгоград']
    
    city = raw_string.split()[0]
    
    if city == 'Москва' or city == 'Санкт-Петербург':
        return city
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'


def get_move_ability(raw_string: str) -> bool:
    """Get move ability from input string

    Args:
        raw_string (str): input data

    Returns:
        bool: true -- can move, false -- else
    """
    
    if ('не готов к перее' in raw_string) or ('не готова к перее' in raw_string):
        return False
    elif 'перее' in raw_string:
        return True
    else:
        return False
        

def get_trip_ability(raw_string: str) -> bool:
    """Get trip ability from input string

    Args:
        raw_string (str): input data

    Returns:
        bool: true -- can get buisness trip, false -- else
    """

    raw_list = raw_string.split(',')[1:]

    for substr in raw_list:
        if 'командировк' in substr and 'готов' in substr:
            return substr.split()[0] != 'не'
    
    return False


proc_df['Город'] = proc_df['Город, переезд, командировки'].apply(get_city)

proc_df['Готовность к переезду'] = proc_df['Город, переезд, командировки'].apply(get_move_ability)

proc_df['Готовность к командировкам'] = proc_df['Город, переезд, командировки'].apply(get_trip_ability)

proc_df = proc_df.drop(['Город, переезд, командировки'], axis=1)
```

```python
# Import exchange currency data
exchange_df = pd.read_csv('/home/ome123/Рабочий стол/DS/Projects/HH_EDA_ANA/Data/ExchangeRates.csv',
                          usecols=['currency', 'date', 'proportion', 'close'])


proc_df['Обновление резюме'] = pd.to_datetime(proc_df['Обновление резюме']).dt.date
exchange_df['date'] = pd.to_datetime(exchange_df['date']).dt.date

currency_name_to_iso_map = {
    'руб': 'RUB',
    'грн': 'UAH',
    'USD': 'USD',
    'EUR': 'EUR',
    'белруб': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}

exchange_df.close /= exchange_df.proportion

# Get salary in RUB via temp_df
temp_df = proc_df[['ЗП', 'Обновление резюме']].copy()
temp_df['currency_val'] = temp_df['ЗП'].str.split().str[0].astype(np.float32)
temp_df['currency_name'] = temp_df['ЗП'].str.split().str[1].str.replace('.', '')
temp_df['currency_name'] = temp_df['currency_name'].map(currency_name_to_iso_map)

temp_df = pd.merge(
    left=temp_df[['currency_val', 'currency_name', 'Обновление резюме']],
    right=exchange_df[['currency', 'date', 'close', 'proportion']],
    left_on=['Обновление резюме', 'currency_name'],
    right_on=['date', 'currency'],
    how='left'
)

temp_df = temp_df[['currency_val', 'currency_name', 'close']]
temp_df['close'] = temp_df['close'].fillna(1)
temp_df.loc[temp_df['currency_name'] == 'RUB', 'close'] = 1
temp_df['salary'] = temp_df.currency_val * temp_df.close


proc_df['ЗП (руб)'] = temp_df['salary'].copy()
proc_df = proc_df.drop(['ЗП'], axis=1)
```
